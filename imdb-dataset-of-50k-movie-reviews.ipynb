IMDB dataset having 50K movie reviews for natural language processing or Text analytics.
This is a dataset for binary sentiment classification containing substantially more data than previous benchmark datasets. We provide a set of 25,000 highly polar movie reviews for training and 25,000 for testing. So, predict the number of positive and negative reviews using either classification or deep learning algorithms.

```python
#loading dataset with pandas
df = pd.read_csv("/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv")
df
```

```python
#checking what percentage of reviews are positive  and negative
df.sentiment.value_counts()
```

so we have a balance dataset

```python
#insert a column to dataset that represent sentiment as numeric values
df['positive'] = df['sentiment'].apply(lambda x:1 if x == 'positive' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split

#split the dataset to train and test batches
X_train, X_test, y_train, y_test = train_test_split(df['review'], df['positive'], test_size=0.2)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.naive_bayes import MultinomialNB as MNB
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.linear_model import LogisticRegression as LG
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier as KNN 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import classification_report
import time
```

```python
#using MultinomialN for making our pipeline training our dataset
start_time = time.time()
clf_mnb = make_pipeline(
    CountVectorizer(),
    MNB()
)

clf_mnb.fit(X_train, y_train)
finish_time = time.time()

print(classification_report(y_test, clf_mnb.predict(X_test)))
print(f'spending time : {finish_time - start_time}')
```

```python
#using RandomForestClassifier for making our pipeline training our dataset
start_time = time.time()
clf_rfc = make_pipeline(
    CountVectorizer(),
    RFC()
)

clf_rfc.fit(X_train, y_train)
finish_time = time.time()

print(classification_report(y_test, clf_rfc.predict(X_test)))
print(f'spending time : {finish_time - start_time}')
```

```python
#using LogisticRegression for making our pipeline training our dataset
start_time = time.time()
clf_lg = make_pipeline(
    CountVectorizer(),
    LG()
)

clf_lg.fit(X_train, y_train)
finish_time = time.time()

print(classification_report(y_test, clf_lg.predict(X_test)))
print(f'spending time : {finish_time - start_time}')
```

```python
#using KNeighborsClassifier for making our pipeline training our dataset
start_time = time.time()
clf_knn = make_pipeline(
    CountVectorizer(),
    KNN(n_neighbors=1)
)

clf_knn.fit(X_train, y_train)
finish_time = time.time()

print(classification_report(y_test, clf_knn.predict(X_test)))
print(f'spending time : {finish_time - start_time}')
```

LogisticRegression pipeline has best result and it is not that much time consuming
